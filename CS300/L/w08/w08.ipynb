Import library

```python
from sklearn.tree import DecisionTreeClassifier
from keras.datasets import mnist
from matplotlib import pyplot
```

Load the training and testing sets into separate variables

```python
(train_X, train_y), (test_X, test_y) = mnist.load_data()
```

Print the shapes of all the vectors

```python
print('X_train: ' + str(train_X.shape))
print('Y_train: ' + str(train_y.shape))
print('X_test:  '  + str(test_X.shape))
print('Y_test:  '  + str(test_y.shape))
```

Plot out dataset

```python
for i in range(9):  
    pyplot.subplot(330 + 1 + i)
    pyplot.imshow(train_X[i], cmap=pyplot.get_cmap('gray'))
    pyplot.show()
```

Create new model

```python
model = DecisionTreeClassifier()
```

Since DecisionTree only take in 2D array, we need to reshape train_X into 2D array by merging its second and third dimension (pixel data)

```python
train_X_2 = train_X.reshape(train_X.shape[0], -1)
train_y_2 = train_y.reshape(-1, 1)
test_X_2 = test_X.reshape(test_X.shape[0], -1)
test_y_2 = test_y.reshape(-1, 1)

print('X_train_2: ' + str(train_X_2.shape))
print('Y_train_2:  '  + str(train_y_2.shape))
print('X_test_2: ' + str(test_X_2.shape))
print('Y_test_2:  '  + str(test_y_2.shape))
```

Fit data into model

```python
model.fit(train_X_2, train_y_2)
```

Test model

```python
correct_pred = 0

for i in range(9):  
    pyplot.subplot(330 + 1 + i)
    pyplot.imshow(test_X[i], cmap=pyplot.get_cmap('gray'))
    prediction = (model.predict(test_X_2[i].reshape(1, -1)))[0]
    actual = test_y_2[i][0]
    is_correct = False
    if str(prediction) == str(actual):
        correct_pred += 1
        is_correct = True
    print("Predicted:", prediction, "| Actual:", actual, "|", "Correct" if is_correct else "Incorrect")
    pyplot.show()
```

```python
score = model.score(test_X_2, test_y_2)
print("Model's score:", score)
```
